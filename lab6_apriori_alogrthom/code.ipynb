```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df=pd.read_csv("../datasets/TVShows_new.csv")
df
```

```python
df['TVShow'] = df['TVShow'].str.strip()
df['TVShow'] = df['TVShow'].str.lower()
```

```python
sumNumVotes=df.groupby("TVShow")["numVotes"].sum()
print(sumNumVotes.values)
```

```python
plt.figure(figsize=(15,5))
sns.barplot(x = sumNumVotes.index[2500:2520], y = sumNumVotes.values[2500:2520], palette = 'gnuplot')
plt.xlabel('TVShow', size = 15)
plt.xticks(rotation=45)
plt.ylabel('Count number Votes', size = 15)
plt.title('Top 20 TV Show purchased by votes', color = 'green', size = 20)
plt.show()
```
